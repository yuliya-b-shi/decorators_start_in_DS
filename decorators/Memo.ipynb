# Декоратор кеширования memo

```python
import functools


def memo(func):
    cache = {}

    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        key = args
        if key in cache:
            return cache[key]
        else:
            result = func(*args, **kwargs)
            cache[key] = result
            return result
    return wrapper

start = time.time()
@memo
def fibonacci(n):
 if n < 2:
  return n
 return fibonacci(n-2) + fibonacci(n-1)
end = time.time()
print("Время работы рекурсивной функции с кешированием memo:", round(end - start, 5))
```

# Сравним эффективность своего декоратора с библиотечным вариантом

```python
import functools
start = time.time()
@functools.lru_cache()
def fibonacci(n):
 if n < 2:
  return n
 return fibonacci(n-2) + fibonacci(n-1)
end = time.time()
print("Время работы рекурсивной функции с кешированием lru_cache:", round(end - start, 5))
```
